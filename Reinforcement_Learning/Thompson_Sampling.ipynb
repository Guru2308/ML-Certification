```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
import random
N = 500
d = 10
ads_selected = []

#STEP-1: Initialize two variable to hold rewards of 1 & 0
no_of_rewards_1 = [0] * d
no_of_rewards_0 = [0] * d

total_reward = 0
for n in range(N):
  ad = 0
  max_sample = 0
  for i in range(d):

    #STEP-2: Random draw from distribution
    random_beta = random.betavariate(no_of_rewards_0[i] +1, no_of_rewards_1[i] +1)

    #STEP-3: Pick the maximum betavariate
    if (random_beta > max_sample):
      max_sample = random_beta
      ad = i

  #Update parameters
  ads_selected.append(ad)
  reward = df.values[n ,ad]
  if reward == 1:
    no_of_rewards_1[ad] += 1
  else:
    no_of_rewards_0[ad] += 1

  total_reward += reward
```

```python
plt.hist(ads_selected)
plt.title("Histogram of the Ads Selection")
plt.xlabel("Ads")
plt.ylabel("No of times ad was selected")
plt.show()
```
