```python
df = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
transactions = []
for i in range(0, 7501):
  #Apriori methods only takes input as LIST not as df.
  # the items in transactions must be STRING always.
  transactions.append([str(df.values[i,j]) for j in range(20)])
```

```python
from apyori import apriori
#min_support = 3*7/7501 => 3-products frequency; 7-days a week;
rules = apriori(transactions=transactions, min_support=0.003, min_lift=3, min_confidence=0.2, min_length=2, max_length=2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
